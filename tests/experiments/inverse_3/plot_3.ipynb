```python
import matplotlib.pyplot as plt
import csv
```

```python
def read_csv(filename: str, dtype=float):
    with open(filename, newline='') as csvfile:
        reader = csv.reader(csvfile, delimiter=',')
        first = True
        lst = []
        rows = 0
        for row in reader:
            if first:
                rows = len(row)
                for i in range(rows):
                    lst.append([])
                first = False

            for j in range(rows):
                lst[j].append(float(row[j]))
        return lst

def get_error(data_1,data_2):
    error = []
    for i in range(len(data_1)):     
        error.append(abs(data_2[i] - data_1[i]))
    return error
```

```python
foot_coordinates = read_csv('output_data/gimbal_extend.csv')
matlab_foot_coordinates = read_csv('matlab_output/gimbal_extend.csv')
```

```python
foot_coordinates_x = foot_coordinates[0]
foot_coordinates_y = foot_coordinates[1]
foot_coordinates_z = foot_coordinates[2]

matlab_foot_coordinates_x = matlab_foot_coordinates[1]
matlab_foot_coordinates_y = matlab_foot_coordinates[2]
matlab_foot_coordinates_z = matlab_foot_coordinates[3]

#time = matlab_foot_coordinates[0]

error_x = get_error(matlab_foot_coordinates_x,foot_coordinates_x)
error_y = get_error(matlab_foot_coordinates_y,foot_coordinates_y)
error_z = get_error(matlab_foot_coordinates_z,foot_coordinates_z)
```

```python
plt.plot(foot_coordinates_x,label="t1")
plt.plot(foot_coordinates_y,label="t2")
plt.plot(foot_coordinates_z,label="e")
plt.plot(matlab_foot_coordinates_x,label="m_t1")
plt.plot(matlab_foot_coordinates_y,label="m_t2")
plt.plot(matlab_foot_coordinates_z,label="m_e")


plt.title("Exp_1_inverse Swing and extend")
plt.legend(loc='upper right')
plt.ylim(-1,1)
plt.show()
```

```python
plt.plot(error_x,label="error_t1")
plt.plot(error_y,label="error_t2")
plt.plot(error_z,label="error_e")
plt.title("Exp_1 Error")
plt.legend(loc='upper right')
plt.show()
```
